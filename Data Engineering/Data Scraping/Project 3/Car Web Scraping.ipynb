### Put everything together inside a For-Loop

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
import urllib.parse
from selenium import webdriver
from selenium.webdriver.support.select import Select
import time


car_name = []
car_price = []
dealer_name = []
review_rating = []
review_count = []
mileage = []

#Open the website
driver = webdriver.Edge(executable_path= r'C:\Users\kana\Documents\Portofolio\Python\Data Scraping\Web Driver\edgedriver_win64\msedgedriver.exe')
driver.get(website)
time.sleep(1)

type_car = Select(driver.find_element_by_xpath('//*[@id="make-model-search-stocktype"]'))
type_car.select_by_visible_text("New & used cars")

type_car = Select(driver.find_element_by_xpath('//*[@id="makes"]'))
type_car.select_by_visible_text("Mercedes-Benz")

search = driver.find_element_by_xpath('//*[@id="by-make-tab"]/div/div[7]/button')
search.click()

for i in range (1,11):
    
    #Get the url
    current_url = driver.current_url

    #store the website params
    headers = ({'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'})

    #Get request
    response = requests.get(current_url, headers=headers)

    soup = BeautifulSoup(response.content, 'html.parser')
    results = soup.find_all('div', {'class':'vehicle-card'})

    for result in results:
        
        try:
            car_name.append(result.find('h2').get_text())
        except:
            car_name.append('')
        
        try:
            car_price.append(result.find('span', {'class':'primary-price'}).get_text().replace('$','').replace(',',''))
        except:
            car_price.append('')
        
        try:
            dealer_name.append(result.find('div', {'class':'dealer-name'}).get_text().strip())
        except:
            dealer_name.append('')
        
        try:
            review_rating.append(result.find('span', {'class':'sds-rating__count'}).get_text())
        except:
            review_rating.append('')
        
        try:
            review_count.append(result.find('span', {'class':'sds-rating__link sds-button-link'}).get_text().replace('(','').replace(')','').replace('reviews', '').replace(',', ''))
        except:
            review_count.append('')
            
        try:
            mileage.append(result.find('div', {'class':'mileage'}).get_text().replace(',','').replace(' mi.',''))
        except:
            mileage.append(0)


    button_next = driver.find_element_by_xpath("//*[@id='next_paginate']")
    button_next.click()
    time.sleep(3)
```

### Create Pandas Dataframe

```python
# dictionary
car_data = pd.DataFrame({'Name': car_name, 'Mileage':mileage, 'Dealer Name':dealer_name,
                                'Rating': review_rating, 'Review Count': review_count, 'Price': car_price})
```
